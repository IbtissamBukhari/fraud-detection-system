# IEEE-CIS Fraud Detection

## Notebook 1: Data Cleaning

This notebook loads the raw transaction and identity data, examines its structure,
handles missing values, fixes data types, and produces a clean dataset ready for
exploratory analysis.

**Dataset:** IEEE-CIS Fraud Detection (Kaggle)  
**Rows:** ~590,000 transactions  
**Target:** isFraud (1 = fraud, 0 = legitimate)

```python
import pandas as pd
import numpy as np

# suppress pandas warnings that clutter the output
pd.set_option('display.max_columns', 100)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

## 1. Load the Data

We load the transaction and identity tables separately and check their shapes
before doing anything else. This tells us how many rows and columns we are
working with.

```python
transaction = pd.read_csv('/kaggle/input/competitions/ieee-fraud-detection/train_transaction.csv')
identity = pd.read_csv('/kaggle/input/competitions/ieee-fraud-detection/train_identity.csv')

df = transaction.merge(identity, on='TransactionID', how='left')

print(f'Transaction table : {transaction.shape}')
print(f'Identity table    : {identity.shape}')
print(f'Merged table      : {df.shape}')
```

```python
df.head()
```

## 2. Memory Optimization

The merged dataframe occupies 2.5 GB using default int64 and float64 types.
Casting to int32 and float32 halves memory usage with no loss of precision
for this dataset.

```python
df.info(memory_usage='deep')
```

```python
# float64 uses 8 bytes per value — float32 uses 4 bytes with sufficient precision for this data
for col in df.select_dtypes(include='float64').columns:
    df[col] = df[col].astype('float32')

# check how much memory we saved
for col in df.select_dtypes(include='int64').columns:
    df[col] = df[col].astype('int32')

print(f'Memory usage after optimization: {df.memory_usage(deep=True).sum() / 1e9:.2f} GB')
```

## 3. Missing Values

414 out of 434 columns have missing values. Identity columns are missing for
transactions where no identity data was collected — this is expected.

Columns missing more than 90% of values are dropped. For the rest, numerical
columns are filled with the median and categorical columns with 'unknown'.

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
}).query('missing_count > 0').sort_values('missing_pct', ascending=False)

print(f'Columns with missing values: {len(missing_df)} out of {df.shape[1]}')
print(missing_df.head(80))
```

```python
# columns missing more than 90% of values carry no meaningful signal
cols_to_drop = missing_df[missing_df['missing_pct'] > 90].index.tolist()

df.drop(columns=cols_to_drop, inplace=True)

print(f'Columns dropped : {len(cols_to_drop)}')
print(f'Remaining columns: {df.shape[1]}')
```

```python
df.info(memory_usage='deep')
```

```python
fraud_counts = df['isFraud'].value_counts()
fraud_pct = df['isFraud'].value_counts(normalize=True) * 100

print(fraud_counts)
print()
print(f'Legitimate transactions: {fraud_pct[0]:.2f}%')
print(f'Fraudulent transactions: {fraud_pct[1]:.2f}%')
```

```python
# numerical columns: fill with median because median is robust to outliers
# categorical columns: fill with the string 'unknown' to preserve the fact that data was missing

num_cols = df.select_dtypes(include=['float32', 'int32']).columns
cat_cols = df.select_dtypes(include='object').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df[cat_cols] = df[cat_cols].fillna('unknown')

print(f'Remaining missing values: {df.isnull().sum().sum()}')
```

## 4. Data Type Validation

We verify that no numeric column contains hidden string values and no categorical
column contains numeric values stored as strings.

```python
print(df.dtypes.value_counts())
print()
print('Object columns:')
print(df.select_dtypes(include='object').columns.tolist())
```

```python
for col in df.select_dtypes(include='object').columns:
    print(f'{col}: {df[col].unique()[:5]}')
```

```python
# check if any object column contains numeric values hiding as strings
for col in df.select_dtypes(include='object').columns:
    try:
        converted = pd.to_numeric(df[col], errors='coerce')
        non_null_pct = converted.notna().sum() / len(df) * 100
        if non_null_pct > 50:
            print(f'{col}: {non_null_pct:.1f}% of values are numeric — consider converting')
    except:
        pass

print('Check complete')
```

```python
# read a small sample of the raw files as plain strings to catch any numeric column hiding string values
raw_sample = pd.read_csv(
    '/kaggle/input/competitions/ieee-fraud-detection/train_transaction.csv',
    nrows=5,
    dtype=str
)

for col in raw_sample.columns:
    vals = raw_sample[col].unique()
    # if a column has non-numeric non-null values it was never truly numeric
    for v in vals:
        try:
            float(v)
        except (ValueError, TypeError):
            if v != 'nan':
                print(f'{col}: contains string value -> "{v}"')
                break
```

```python
df.to_csv('cleaned_data.csv', index=False)

print(f'Saved: cleaned_data.csv')
print(f'Shape: {df.shape}')
```
